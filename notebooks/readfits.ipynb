```python
from astropy.io import fits
from matplotlib import pyplot as plt
import numpy as np
spec=fits.open("berk_composte.fits")
hdu0=spec[0]
d=hdu0.data
w=d[0]
f=d[1]

fig,ax=plt.subplots(figsize=(9,5))
ax.set_xscale("log")
ax.set_yscale("log")
ax.plot(w,f)
# array indexing：select the pixels you want
idx=((w>=1350) & (w<=1360)) | ((w>=4200) & (w<=4230))
ax.scatter(w[idx],f[idx],color='red')
#Model fitting: help(np.polyfit)

a=np.polyfit(np.log10(w[idx]),np.log10(f[idx]),1) #y=ax+b
#calculate the fitted values
log10_f1=np.log10(w)*a[0]+a[1]
f1=10**log10_f1
# let's play with the fitted parameters by changing them a bit.
log10_f2=np.log10(w)*(a[0]+0.1)+a[1]-0.35
log10_f3=np.log10(w)*(a[0]-0.1)+a[1]+0.35
f2=10**log10_f2
f3=10**log10_f3
#explore various artist setting using: help(ax.plot)
#different models 
ax.plot(w,f1,'b-',linestyle='dashed')
ax.plot(w,f2,'r-',linestyle='dashed')
ax.plot(w,f3,'g-',linestyle='dashed')

plt.ylim(1,25)
plt.xticks(np.arange(1,10)*1000,np.arange(1,10)*1000)
plt.yticks([0.5,1,5,10],[0.5,1,5,10])
ax.set_xlabel(r"$\mathrm{Wavelength (\AA)}$")
ax.set_ylabel("Flux")
ax.text(1000,1,r"$\alpha_\lambda=-1.4$")

#c.f. https://matplotlib.org/stable/tutorials/text/annotations.html
ax.annotate(r'H$\alpha$', xy=(6563, 5),  xycoords='data',
            xytext=(6563,10), textcoords='data',
            #you could try arrow style here
            arrowprops=dict(arrowstyle='-',facecolor='green') ,
            horizontalalignment='center', verticalalignment='top',
            )
plt.show()
```

## lmfit核心概念: Parameter和Parameters类型

```python
from lmfit.models import PowerLawModel,GaussianModel
from lmfit import Parameter,Parameters
```

```python
#lmfit内置了很多的模型：https://lmfit.github.io/lmfit-py/builtin_models.html
pl_mod=PowerLawModel(prefix='pl_')
# 注意这里的参数pars仅仅是power law模型的参数
pars=pl_mod.guess(f[idx],x=w[idx]) #注意这里的idx是连续谱窗口的数据点
gau_mod=GaussianModel(prefix='g1_')
```

```python
#注意pars目前仅仅power law的变量
#https://lmfit.github.io/lmfit-py/parameters.html
pars
```

```python
#定义模型
mod=pl_mod+gau_mod
```

```python
pars
```

```python
# 根据需要，我们可以设置让某个参数在拟合时不变化
pars['pl_amplitude'].set(vary=False)
pars['pl_exponent'].set(vary=False)
```

```python
#把高斯模型的参数update添加到pars中：
pars.update(gau_mod.make_params())
```

```python
# 根据需要，我们可以设置让某个参数在某个区间内变化
pars['g1_center'].set(value=1549,min=1546,max=1553)
pars['g1_sigma'].set(value=10,min=5,max=20)
pars
```

```python
#设置完各种参数的初始值和限制后，真正的拟合在这里
out=mod.fit(f,pars,x=w)
```

```python
#画出拟合的结果:residual(残差)和best fit(最佳拟合)
out.plot()
```

```python
out.params
```

```python
#根据拟合出来的参数，计算出模型拟合的结果
f_model=out.eval(x=w)
```

```python
out.params
```

```python
pl_pars=Parameters()
pl_pars.add_many(out.params['pl_amplitude'],out.params['pl_exponent'])
pl_model=pl_mod.eval(pl_pars,x=w)
pl_pars
```

```python
gau_pars
```

```python
gau_pars=Parameters()
gau_pars.add_many(out.params['g1_amplitude'],out.params['g1_center'],out.params['g1_sigma'])
gau_model=gau_mod.eval(gau_pars,x=w)
plt.plot(w,gau_model)
```

```python
#另外一种计算模型结果的方法
gau_model1=gau_mod.eval(out.params,x=w)
plt.plot(w,gau_model1)
```

```python
%matplotlib inline
fig,ax=plt.subplots(figsize=(9,5))
ax.set_xscale("log")
ax.set_yscale("log")
ax.plot(w,f)
#模型拟合的结果再这里
ax.plot(w,f_model,'b-',linestyle='dashed')
ax.plot(w,pl_model,'r-',linestyle='dashed')
ax.plot(w,gau_model,'g-',linestyle='dashed')
plt.ylim(1,25)
plt.xticks(np.arange(1,10)*1000,np.arange(1,10)*1000)
plt.yticks([0.5,1,5,10],[0.5,1,5,10])
ax.set_xlabel(r"$\mathrm{Wavelength (\AA)}$")
ax.set_ylabel("Flux")
ax.text(1000,1,r"$\alpha_\lambda=-1.4$")

#c.f. https://matplotlib.org/stable/tutorials/text/annotations.html
ax.annotate(r'H$\alpha$', xy=(6563, 5),  xycoords='data',
            xytext=(6563,10), textcoords='data',
            #you could try arrow style here
            arrowprops=dict(arrowstyle='-',facecolor='green') ,
            horizontalalignment='center', verticalalignment='top',
            )
plt.show()
```

## Using Mathematical Constraints in lmfit
https://lmfit.github.io/lmfit-py/constraints.html

```python
pars = Parameters()
pars.add('amp_1', value=0.5, min=0, max=1)
pars.add('cen_1', value=2.2)
pars.add('wid_1', value=0.2)
pars
```

```python
#constraint here
#通过设置constraint，我们可以绑定两个变量
pars.add('amp_2', expr='(2.0 - amp_1**2)')
pars.add('wid_2', expr='sqrt(pi)*wid_1')
pars.add('cen_2', expr='cen_1 * wid_2 / max(wid_1, 0.001)')
pars
```

```python
pars.add('param_a', value=1)
pars.add('param_b', value=2)
pars.add('test_val', value=100)

pars.add('bounded', expr='param_a if test_val/2. > 100 else param_b')
pars
```

## 模型：另外一个lmfit的核心概念
* https://lmfit.github.io/lmfit-py/model.html 
* built-in models: https://lmfit.github.io/lmfit-py/builtin_models.html
* lmfit自定义模型拟合

```python
import matplotlib.pyplot as plt
from numpy import exp, loadtxt, pi, sqrt

from lmfit import Model

data = loadtxt('model1d_gauss.dat')
x = data[:, 0]
y = data[:, 1]


def gaussian(x, amp, cen, wid):
    """1-d gaussian: gaussian(x, amp, cen, wid)"""
    return (amp / (sqrt(2*pi) * wid)) * exp(-(x-cen)**2 / (2*wid**2))

#注意这里的函数定义：
#第一个参数（这里的x)默认是自变量
#后面的参数和关键字（这里的amp,cen,wid)默认为模型的参数

```

```python
plt.plot(x,y)
```

```python
gmodel = Model(gaussian)
```

```python
gmodel?
```

```python
result = gmodel.fit(y, x=x, amp=5, cen=5, wid=1)

print(result.fit_report())
```

```python
result.init_fit
```

```python
plt.plot(x, y, 'o')
plt.plot(x, result.init_fit, '--', label='initial fit')
plt.plot(x, result.best_fit, '-', label='best fit')
plt.legend()
plt.show()
# <end examples/doc_model_gaussian.py>
```

```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
from scipy.signal import savgol_filter
%matplotlib inline
```

```python
hdulist=fits.open('spec-6173-56238-0528.fits')
spec=hdulist[1].data
plt.figure(figsize=(8,5)) #change figure size
wave=10**spec['loglam']
flux=spec['flux']
flux_smooth=savgol_filter(spec['flux'], 51, 3)
flux_model=spec['model']
# plt.plot(10**spec['loglam'],spec['flux'],10**spec['loglam'],flux_smooth,10**spec['loglam'],spec['model'])
plt.plot(wave,flux,wave,flux_smooth,wave,flux_model)
plt.ylim(-1,5)
plt.show()
```

## signal to noise ratio

```python
error=1/np.sqrt(spec['ivar'])
sn=flux/error
plt.plot(wave,sn)
plt.xlim(3600,5000)
# index Ca II region: Ca II 3934,3969
idx=((wave>=3930) & (wave<=3975))
plt.plot(wave[idx],sn[idx],c='r')
```
